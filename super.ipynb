# 상속

```python
class Person():
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def get_name(self):
    print(f'name :{self.name}')

  def get_age(self):
    print(f'age : {self.age}')
```

```python
class Student(Person):
  def __init__(self, name, age, GPA):
    super().__init__(name, age)
    self.GPA = GPA
    
  def get_name(self):
        print(f'저는 대학생 {self.name}입니다.')

  def get_GPA(self):
        super().get_name()
        print(f'제 학점은 {self.GPA}입니다.')
```

```python
st = Student('h', 33, 123)
st.get_age()
st.get_name()
st.get_GPA()
```
